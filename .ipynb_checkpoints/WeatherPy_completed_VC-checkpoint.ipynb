```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import openweather_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#create sample subset for testing purposes (Only used for troubleshooting and code design)
sampleset = []
index = 0

for city in cities:
    sampleset.append(city)
    index = index + 1
    if index == 75:
        break
```

```python
#Base_URL for query
base_url = "http://api.openweathermap.org/data/2.5/weather?"

#Instantiate weatherdata as an empty list and parameters
weatherdata = []
index = 0
setnum = 1

print("-----------------------------------------------")
print("Initiate Data Retrieval...")
print("-----------------------------------------------")

#Iterate across list of cities and get list of JSONs
for city in cities:
    #Cobble together query_url and send request
    query_url = f"{base_url}appid={openweather_key}&q={city}&units=imperial"
    weather_response = requests.get(query_url)
    
    #Ping the server in sets of 60, stop every 60 queries for 60 seconds because the API documentation
    #states that a free account may perform 60 queries a minute. If it comes back with data, append it to 
    #the weatherdata list. Otherwise, print that the city was not found and continue.
    if weather_response.status_code == 200:
        weatherdata.append(weather_response.json())
        print(f"Retrieving Record {index} of Data Set {setnum} information | {city}")
        index = index + 1
        if index%60 == 0:
            index = 0
            setnum = setnum + 1
            if(setnum != 1):
                print("Wait 60 seconds")
                time.sleep(60)
    elif weather_response.status_code == 404:
        print(f"City not found for Record {index} of Set {setnum}. Skipping | {city}")
        index = index + 1
print("-----------------------------------------------")
print("Data Retrieval Complete")
print("-----------------------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Create Lists for Each Column in DataFrame
name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
maxtemp = []
windspeed = []

#Populate Lists for Each Column in DataFrame
for i in range(len(weatherdata)):
    name.append(weatherdata[i]['name'])
    cloudiness.append(weatherdata[i]['clouds']['all'])
    country.append(weatherdata[i]['sys']['country'])
    date.append(weatherdata[i]['dt'])
    humidity.append(weatherdata[i]['main']['humidity'])
    lat.append(weatherdata[i]['coord']['lat'])
    lng.append(weatherdata[i]['coord']['lon'])
    maxtemp.append(weatherdata[i]['main']['temp_max'])
    windspeed.append(weatherdata[i]['wind']['speed'])
```

```python
#Create Dictionary Based on Populated List
weatherdict = {'City':name, 
               'Cloudiness':cloudiness,
               'Country':country,
               'Date':date,
               'Humidity':humidity,
               'Latitude':lat,
               'Longitude':lng,
               'Max Temp':maxtemp,
               'Wind Speed':windspeed}

#Use Pandas DataFrame to create new df
df = pd.DataFrame(weatherdict)
df.head()
```

```python
#Count the number of overall records
df.count()
```

```python
#Write weatherData to a csv file
df.to_csv("./weatherData.csv",index=False)
```
